# Feature engineering

```python
## Converting categories into numerical
for col in df.dtypes[df.dtypes == 'object'].index:
    for_dummy = df.pop(col)
    df = pd.concat([df,pd.get_dummies(for_dummy,prefix = col)],axis = 1)
```

```python
df
```

# Split data into X and Y

```python
X = df.drop('expenses',axis =1)
y = df['expenses']
```

```python
X
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 77)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn import metrics
```

```python
model.coef_
```

```python
model.intercept_
```

```python
print('r2_score:',r2_score(y_test,y_pred))
print('mean_score:',mean_squared_error(y_test,y_pred))
print('meab ab_score:',mean_absolute_error(y_test,y_pred))
```

```python
print("rtmean:", np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
model.score(X_train,y_train)
```

```python
dframe  = pd.DataFrame({'Actual':y_test,"predict": y_pred,"varince": y_test-y_pred})
dframe
```
